# Our method vs SEFR

```python
!wget https://github.com/PyThaiNLP/wisesight-sentiment/raw/master/word-tokenization/wisesight-1000-samples-tokenised.label
```

```python
engine=['deepcut','ws1000']
sefr_cut.load_model(engine='deepcut')
```

```python
def cutcut(text):
    cut_list = []
    sefr_cut.load_model(engine='deepcut')
    cut_list.append(['|'.join(sefr_cut.tokenize(text)[0]),'deepcut'])
    sefr_cut.load_model(engine='ws1000')
    cut_list.append(['|'.join(sefr_cut.tokenize(text)[0]),'sefr-ws1000'])
    
    oskut.load_model(engine='ws')
    cut_list.append(['|'.join(oskut.OSKut(text)),'mde-ws'])
    oskut.load_model(engine='ws-augment-60p')
    cut_list.append(['|'.join(oskut.OSKut(text)),'mde-ws-augment-60p'])
    
    return cut_list
```

```python
with open('wisesight-1000-samples-tokenised.label','r',encoding='utf-8-sig') as f:
    file = f.readlines()
np.random.seed(42)
random_list = np.random.randint(len(file), size=50)
```

```python
cut_list = []
for item in random_list:
    cut_return = cutcut(file[item].replace('|',''))
    cut_list.append(cut_return)
```

```python
for idx,item in enumerate(random_list):
    print(f'mode: Answer\ntext:{file[item]}')
    [print(f'mode:{i[1]}\ntext:{i[0]}') for i in cut_list[idx]]
    print('#################################################')
```
